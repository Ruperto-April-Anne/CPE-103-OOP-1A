Create a Python Class for Student

```python
class grades():
  def __init__(self, course, student_id, name, program, sex, weight, height):
    self.course = course
    self.student_id = student_id
    self.name = name
    self.program = program
    self.sex = sex
    self.weight = weight
    self.height = height

  def compute_cs(self,hands_on, quiz, assignment):
    return (hands_on * 0.5) + (quiz * 0.3) + (assignment * 0.2)

  def compute_pg(self, prelim_exam, prelim_hands_on, prelim_quiz, prelim_assignment):
    prelim_cs = self.compute_cs(prelim_hands_on, prelim_quiz, prelim_assignment)
    return (prelim_exam * 0.50) + (prelim_cs * 0.50)


  def compute_mg(self, prelim_grade, midterm_exam, midterm_hands_on, midterm_quiz, midterm_assignment):
    midterm_cs = self.compute_cs(midterm_hands_on, midterm_quiz, midterm_assignment)
    midterm_avg = (midterm_exam * 0.50) + (midterm_cs * 0.50)
    return (prelim_grade * 1/3) + (midterm_avg * 2/3)

  def compute_fg(self, midterm_grade, final_exam, final_hands_on, final_quiz, final_assignment):
    final_cs = self.compute_cs(final_hands_on, final_quiz, final_assignment)
    final_avg = (final_exam * 0.50) + (final_cs * 0.50)
    return (midterm_grade * 1/3) + (final_avg * 2/3)

  def grades_input(self):
    self.prelim_exam = float(input("\nEnter Prelim Exam Score: "))
    self.prelim_hands_on = float(input("Enter Prelim Hands-on Score: "))
    self.prelim_quiz = float(input("Enter Prelim Quiz Score: "))
    self.prelim_assignment = float(input("Enter Prelim Assignment Score: "))
    self.midterm_exam = float(input("\nEnter Midterm Exam Score: "))
    self.midterm_hands_on = float(input("Enter Midterm Hands-on Score: "))
    self.midterm_quiz = float(input("Enter Midterm Quiz Score: "))
    self.midterm_assignment = float(input("Enter Midterm Assignment Score: "))
    self.final_exam = float(input("\nEnter Final Exam Score: "))
    self.final_hands_on = float(input("Enter Final Hands-on Score: "))
    self.final_quiz = float(input("Enter Final Quiz Score: "))
    self.final_assignment = float(input("Enter Final Assignment Score: "))
    self.prelim_grade = self.compute_pg(self.prelim_exam, self.prelim_hands_on, self.prelim_quiz, self.prelim_assignment)
    self.midterm_grade = self.compute_mg(self.prelim_grade, self.midterm_exam, self.midterm_hands_on, self.midterm_quiz, self.midterm_assignment)
    self.final_grade = self.compute_fg(self.midterm_grade, self.final_exam, self.final_hands_on, self.final_quiz, self.final_assignment)

  def print_grades(self):
    print("\nRESULTS")
    print(f"Prelim Grade: {self.prelim_grade:.2f}")
    print(f"Midterm Grade: {self.midterm_grade:.2f}")
    print(f"Final Grade: {self.final_grade:.2f}")

class student(grades):
  pass

student_list = []
for i in range(3):
  print(f"\nEnter details for Student {i + 1}:")
  name = input("Enter your name: ")
  student_id = input("Enter your Student ID: ")
  course = print("Course: CPE103")
  sex = input("Sex: ")
  weight = float(input("Weight(kg): "))
  height = float(input("Height(cm): "))
  student_instance = student(course, student_id, name, course, sex, weight, height)
  student_instance.grades_input()
  student_instance.print_grades()
```
